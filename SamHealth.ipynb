```python
heart_rate.drop(['source','tag_id','heart_beat_count','custom','binning_data','heart_rate.max','heart_rate.min','comment','datauuid'], axis=1, inplace=True)
```

```python
heart_rate.head()
```

```python
heart_rate['update_time'] = heart_rate['update_time'].dt.date
```

```python
heart_stats = heart_rate.groupby('update_time')['heart_rate'].mean()
heart_stats = heart_stats.to_frame()
```

```python
heart_stats.head()
```

```python
heart_max = heart_rate.groupby('update_time')['heart_rate'].max()
heart_max.head()
```

```python
heart_min = heart_rate.groupby('update_time')['heart_rate'].min()
heart_min.head()
```

```python
heart_stats = heart_stats.merge(heart_max, on='update_time')
heart_stats = heart_stats.merge(heart_min, on='update_time')
```

```python
heart_stats.head()
```

```python
heart_stats.rename(columns={'heart_rate_x':'average_rate',
                           'heart_rate_y':'max_rate',
                           'heart_rate':'min_rate'}, inplace=True)
```

```python
heart_stats.reset_index(inplace=True)
heart_stats.head()
```

```python
heart_stats.info()
```

```python
heart_stats['update_time'] = heart_stats['update_time'].apply(pd.to_datetime)
```

```python
heart_stats.info()
```

```python
heart_stats.head()
```

```python
heart_fig = px.line(heart_stats, x='update_time', y='average_rate', title='Average Heart Rate Overtime', height = 500)
heart_fig.show()
```

This graph lacks data in the beginning, and that is because I started wearing a smartwatch that constantly recorded my heartrate. Before this, I only took my heart rate sparsely. Because of this everyting before the time I started using a smartwatch is unimportant, so I want to filter that period out. 

```python
#looked back on my Amazon orders so I know that I started wearing it on either May 28th or 29th. 29th it is.
heart_stats[heart_stats['update_time'] >= pd.to_datetime('2022-05-29')]
```

```python
heart_stats = heart_stats[heart_stats['update_time'] >= pd.to_datetime('2022-05-29')]
```

```python
heart_fig = px.line(heart_stats, x='update_time', y='average_rate', title='Average Heart Rate Overtime', height = 500)
heart_fig.show()
```

Pretty consistent heartbeat, except for those weird anomalies. Maybe those where days I just wore my watch during a run.
It seems that my average is in the 60s, with some being in the lows, and some in the highs. I would like to get my resting heart rate, as I feel that is the most important metric, however that stat was not available. I checked on my phone and apparently my resting heart rate is 56 bpm

```python
heart_fig_max = px.line(heart_stats, x='update_time', y='max_rate', title='Max Heart Rate Overtime', height = 500)
heart_fig_max.show()
```

Yeah my heart rate is way more irratic than the average as expected. Most likely due the days when I am active and the days I am not.
