```python
def line_overtime(df, cols):
    for col in cols:
        fig = px.line(df, x='update_time', y=col, title=f'{col} Overtime', height = 500)
        fig.show()

def box_overall(df, cols, title):
    fig = px.box(df, y=cols,title=title, height = 700)
    fig.show()

def box_overtime(df, cols):
    temp = df
    temp['month_year'] = pd.to_datetime(temp['update_time']).dt.strftime('%Y-%m')
    for col in cols:
        fig = px.box(temp, x='month_year',y=col,title=f'Monthly {col} distribution(Box)', height = 700)
        fig.show()
    
```

```python
line_overtime(General_exer, ['duration','mean_heart_rate','max_heart_rate','total_calorie'])
```

In these charts, especially in the duration and total calories, there is a noticable upwards trend starting at May. This is mostly likly when I started working out longer. The box polts overtime should give some more details.

```python
box_overtime(General_exer, ['duration','mean_heart_rate','max_heart_rate','total_calorie'])
```

The trend starting at May is way more prononced and noticable in May, withit only going higher in the following months. Also, although slightly less noticable, my average heart rate also seems to go in a upwards trend.
